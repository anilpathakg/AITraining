```python
# âœ… Install the required library first (uncomment if needed)
#pip install langgraph

from langgraph.graph import StateGraph

# ðŸ”¹ Define the first agent node
def agent_a(state):
    state["a_output"] = "Hello from A"
    print("Agent A ran and added:", state["a_output"])
    return state

# ðŸ”¹ Define the second agent node
def agent_b(state):
    state["b_output"] = state["a_output"] + ", and B says hi!"
    print("Agent B ran and added:", state["b_output"])
    return state

# ðŸ”¹ Create a LangGraph workflow passing shared state (`dict`)
graph = StateGraph(dict)

# ðŸ”¹ Add agent nodes to the graph
graph.add_node("A", agent_a)  # Node A runs agent_a()
graph.add_node("B", agent_b)  # Node B runs agent_b()

# ðŸ”¹ Set A as the entry point (start of workflow)
graph.set_entry_point("A")

# ðŸ”¹ Connect A â†’ B (define sequence)
graph.add_edge("A", "B")

# ðŸ”¹ Compile graph into an executable app
app = graph.compile()

# ðŸ”¹ Invoke the graph with initial (empty) state
result = app.invoke({})

# ðŸ”¹ Print final state (after all agents have run)
print("\nðŸ”š Final Result:", result)
```
