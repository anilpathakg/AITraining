```python
# ================================================
# 🧩 STEP 1: Install libraries
# ================================================
!pip install -qU sentence-transformers scikit-learn matplotlib

# ================================================
# ⚙️ STEP 2: Import and visualize embeddings
# ================================================
from sentence_transformers import SentenceTransformer
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Load local model (free, no API key)
model = SentenceTransformer("all-MiniLM-L6-v2")

# Sample sentences
texts = [
    "cat",
    "dog",
    "car",
    "apple",
    "banana",
    "safety in factories",
    "worker protection",
    "fruit taste",
    "vehicle speed",
    "pet care",
    "rat",
    "mango"
]

# Generate embeddings
embeddings = model.encode(texts)

# Reduce 384D → 2D
pca = PCA(n_components=2)
points = pca.fit_transform(embeddings)

# Plot
plt.figure(figsize=(8,6))
plt.scatter(points[:, 0], points[:, 1])
for i, word in enumerate(texts):
    plt.annotate(word, (points[i, 0]+0.01, points[i, 1]+0.01))
plt.title("2D Visualization of Text Embeddings (Offline Model)")
plt.xlabel("PCA Dimension 1")
plt.ylabel("PCA Dimension 2")
plt.show()
```
